```python
import pandas as pd
import matplotlib.pyplot as pt
import pandas as pd
#from sklearn import Multiple_Model
```

```python
df=pd.read_csv("multilinear data.csv")
df
```

```python
df["sex"]=df["sex"].astype("category")
df["sex"]=df["sex"].cat.codes
df
```

```python
df["smoker"]=df["smoker"].astype("category")
df["smoker"]=df["smoker"].cat.codes

df["region"]=df["region"].astype("category")
df["region"]=df["region"].cat.codes
df
```

```python
df.isnull().sum()
```

```python
#FORMUL => y=m1x1+m2x2+m3x3............mnxn
```

```python
#defining independent variables
x=df.drop(columns="charges")
x
```

```python
# defing dependent variable
y=df["charges"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred=reg.predict(x_train)
y_pred
```

```python
pt.scatter(y_train,y_pred)
pt.xlabel("actual values")
pt.ylabel("predicted values")
pt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,y_pred)
```

```python
y_test_pred=reg.predict(x_test)
```

```python
pt.scatter(y_test,y_test_pred)
pt.xlabel("actual values")
pt.ylabel("predicted values")
pt.show()
```

```python
r2_score(y_test,y_test_pred)
```
